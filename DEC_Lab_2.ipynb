```python
import pandas as pd
dataset=pd.read_csv("hotel_booking.csv")
print(dataset)
```

```python
dataset.isnull().sum()
```

```python
dataset.dtypes
```

```python
dataset.dropna(axis=0, thresh=2)
# Remove rows with at least 2 NaN values
```

```python
dataset.fillna(0)
 # Replace NaN with 0
```

```python
dataset.drop_duplicates()
# Data Deuplictae removal
```

```python
dataset.drop_duplicates(subset=['reservation_status_date', 'arrival_date_year','arrival_date_month',	'arrival_date_week_number','arrival_date_day_of_month'])
```

```python
dataset.rename(columns={'hotel':'Faxy Muse'})
```

```python
dataset.replace({'Resort Hotel': 'Hell Ya'})
```

```python
dataset['hotel'] = dataset['hotel'].str.strip()
dataset['arrival_date_month'] = dataset['arrival_date_month'].str.lower()
```

```python
dataset['lead_time'] = dataset['lead_time'].astype(float)
```

```python
dataset.dtypes
```

```python
dataset.sort_values(by='lead_time', ascending=False)
```

```python
dataset.sort_values(by='lead_time', ascending=True)
```

```python
dataset[dataset['agent'].isnull()]
```

```python
def custom_function(x):
  return x +100
dataset['adr'] = dataset['adr'].apply(custom_function)
```
