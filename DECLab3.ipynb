```python
# Load the dataset and display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

dataset=pd.read_csv("hotel_booking.csv")

# Identify duplicates
duplicates = dataset.duplicated()

# Handle data redundancy(drop duplicates)
dataset_cleaned = dataset.drop_duplicates()

# Handling missing data (filing data)
dataset_filled = dataset_cleaned.fillna(0)

# Perform Correlation analysis(Pearson's ratio)
numeric_columns = dataset_filled.select_dtypes(include='number')
corr_matrix = numeric_columns.corr(method='pearson')
print("----------Pearson Correlation Matrix:----------")
print(corr_matrix)

# Display heat map
plt.figure(figsize=(13, 9),dpi=230)
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('--------Heat Map--------')
plt.show()

# Data Visualisation - Historgram
plt.figure(figsize=(8, 6))
sns.histplot(numeric_columns['lead_time'], kde=True, bins=30, color='skyblue')
plt.title('Histogram with Density Plot')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Data Visualisation - Bar Graph
plt.figure(figsize=(16, 10))
sns.countplot(x='arrival_date_day_of_month', data=dataset_filled, palette='Set2')
plt.title('Bar Graph (Count of Categories)')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.show()

# Data Visualisation - Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='stays_in_week_nights', y='arrival_date_day_of_month', data=dataset_filled, color='blue', marker='o')
plt.title('Scatter Plot between column_1 and column_2')
plt.xlabel('stays_in_week_nights')
plt.ylabel('arrival_date_day_of_month')
plt.show()

# Perfrom min/max scaling
print("----------Min/Max Scaling:----------")
scaler = MinMaxScaler()
columns_to_scale = dataset_filled[['lead_time', 'adults']]
scaled_data = scaler.fit_transform(columns_to_scale)
scaled_df = pd.DataFrame(scaled_data, columns=['lead_time', 'adults'])
print(scaled_df.head())

# Perform ZScore-scaling
print("----------ZScore-----------")
scaler = StandardScaler()
scaled_data = scaler.fit_transform(columns_to_scale)
scaled_df = pd.DataFrame(scaled_data, columns=['lead_time', 'adults'])
print(scaled_df.head())

# Perform feature smoothing using Binning Method
print("----------Binning Method - Equal Width B -----------")
dataset_fillednew = dataset.fillna(0)
column_to_bin = dataset_fillednew['lead_time']
bins = 5
bin_labels = [f'Bin {i+1}' for i in range(bins)]
binned_column = pd.cut(column_to_bin, bins=bins, labels=bin_labels)
dataset_fillednew['lead_time_binned'] = binned_column
print(dataset_fillednew[['lead_time', 'lead_time_binned']].head())


# Perfrom feature selection to select most significant feature
# Display same using visual tech
```
